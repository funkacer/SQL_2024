```python
import psycopg2
```

```python
conn = psycopg2.connect('''host=localhost port=5432 dbname=test user=funka password=funka1''')
```

```python
with conn.cursor() as cur:
    cur.execute('''
    select * from information_schema.tables where table_schema = 'public'
    ''')
    description = cur.description
    data = cur.fetchall()
```

```python
columns = [c.name for c in description]
```

```python
columns
```

```python
# udělej namedtuple
```

```python
from collections import namedtuple
```

```python
Rows = namedtuple('Row', ' '.join(columns))
```

```python
row = Rows(*data[0])
```

```python
row
```

```python
rows = {}
for index, row in enumerate(data):
    rows[index+1] = Rows(*row)
```
